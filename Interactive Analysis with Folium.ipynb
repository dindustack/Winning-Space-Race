```python
# TASK 3: Calculate the distances between a launch site to its proximities

from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Define the distance calculation function (provided in the notebook)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Earth's radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Initialize the map centered at CCAFS LC-40
launch_site = spacex_df[spacex_df['Launch Site'] == 'CCAFS LC-40'][['Lat', 'Long']].iloc[0]
launch_site_lat = launch_site['Lat']
launch_site_lon = launch_site['Long']
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add MousePosition to the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Define proximity points
proximities = [
    {'name': 'Coastline', 'Lat': 28.56367, 'Long': -80.57163},  # From notebook example
    {'name': 'Cape Canaveral City', 'Lat': 28.3922, 'Long': -80.6077},
    {'name': 'Railway', 'Lat': 28.5600, 'Long': -80.5800},
    {'name': 'Highway SR 405', 'Lat': 28.5620, 'Long': -80.5850}
]

# Add launch site marker
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='CCAFS LC-40',
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(site_map)

# Add proximity markers, distance lines, and labels
for proximity in proximities:
    # Calculate distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, proximity['Lat'], proximity['Long'])
    
    # Add proximity marker
    folium.Marker(
        location=[proximity['Lat'], proximity['Long']],
        popup=f"{proximity['name']} ({distance:.2f} km)",
        icon=folium.Icon(color='orange')
    ).add_to(site_map)
    
    # Add PolyLine
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [proximity['Lat'], proximity['Long']]],
        color='blue',
        weight=2,
        opacity=0.8
    ).add_to(site_map)
    
    # Add distance label at midpoint
    mid_lat = (launch_site_lat + proximity['Lat']) / 2
    mid_lon = (launch_site_lon + proximity['Long']) / 2
    folium.Marker(
        location=[mid_lat, mid_lon],
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(75, 18),
            html=f'<div style="font-size: 12pt">{distance:.2f} km</div>'
        )
    ).add_to(site_map)

# Save or display the map
site_map.save('launch_site_proximities_map.html')
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to the closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
import pandas as pd
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant columns and create launch_sites_df
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()[['Launch Site', 'Lat', 'Long']]

# Initialize the map centered at NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Task 1: Add launch sites with circles and markers
for index, site in launch_sites_df.iterrows():
    folium.Circle(
        location=[site['Lat'], site['Long']],
        radius=1000,
        color='#d35400',
        fill=True
    ).add_child(folium.Popup(site['Launch Site'])).add_to(site_map)
    
    folium.Marker(
        location=[site['Lat'], site['Long']],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{site["Launch Site"]}</b></div>'
        )
    ).add_to(site_map)

# Task 2: Add success/failed launch markers
marker_cluster = MarkerCluster()
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
site_map.add_child(marker_cluster)

for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"{record['Launch Site']} - {'Success' if record['class'] == 1 else 'Failure'}"
    )
    marker_cluster.add_child(marker)

# Task 3: Add MousePosition and proximity markers/lines
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Distance calculation function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Proximity coordinates for CCAFS LC-40 (approximated)
launch_site = {'name': 'CCAFS LC-40', 'Lat': 28.562302, 'Long': -80.577356}
proximities = [
    ('Coastline', {'Lat': 28.56367, 'Long': -80.57163}),
    ('Railway', {'Lat': 28.57205, 'Long': -80.58528}),
    ('Highway', {'Lat': 28.56250, 'Long': -80.57050}),
    ('City', {'Lat': 28.61200, 'Long': -80.80800})  # Titusville
]

# Add markers and lines for proximities
for feature, coords in proximities:
    distance = calculate_distance(launch_site['Lat'], launch_site['Long'], coords['Lat'], coords['Long'])
    # Add a marker for the proximity
    distance_marker = folium.Marker(
        [coords['Lat'], coords['Long']],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{feature}: {distance:.2f} KM</b></div>'
        )
    )
    site_map.add_child(distance_marker)
    
    # Draw a PolyLine between the launch site and the proximity
    lines = folium.PolyLine(
        locations=[[launch_site['Lat'], launch_site['Long']], [coords['Lat'], coords['Long']]],
        weight=1,
        color='blue'
    )
    site_map.add_child(lines)

# Save or display the map
site_map.save('spacex_launch_map.html')
site_map
```
